```python
##Visualise top 5 Districts


# Assuming df is your DataFrame
top_districts = df["district"].value_counts().head(5).index
sns.countplot(x=df["district"], hue=df["district"], palette="YlGnBu", order=top_districts, legend=False)

# Show the plot
plt.show()
```

```python
##Please provide a Pie Diagram for top 5 basis for search


search_basis_split = df['basis_for_search'].str.split('|').explode().str.strip()
basis_counts = search_basis_split.value_counts()
top_5_basis = basis_counts.head(5)
top_5_basis.plot.pie(autopct='%1.1f%%', startangle=90, legend=False)
plt.title('Top 5 Basis for Search')
plt.show()
```

```python

##Which year records most numbers of incidents

df['stop_datetime'] = pd.to_datetime(df['stop_datetime'], format="%d-%m-%Y %H:%M")


incident_counts = df['stop_datetime'].value_counts().sort_index().reset_index()
incident_counts.columns = ['Date', 'Incident Count']


incident_counts.plot.line(x='Date', y='Incident Count', marker='o', linestyle='-')


plt.xlabel('Date')
plt.ylabel('Incident Count')
plt.title('Incident Count Over Time')


plt.show()
```

```python
##What result of stop was maximum help me with top 5

top_results = df['results_of_stop'].value_counts().head(5)
df_top_results = df[df['results_of_stop'].isin(top_results.index)].copy()
df_top_results['results_of_stop'] = pd.Categorical(df_top_results['results_of_stop'], categories=top_results.index, ordered=True)
counts = df_top_results['results_of_stop'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85, wedgeprops=dict(width=0.4, edgecolor='w'))
centre_circle = plt.Circle((0, 0), 0.6, color='white', edgecolor='black', linewidth=0.8)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Top 5 Results of Police Stops (Donut Chart)')
plt.show()
```
